# Control Flow

There are three basic types of control flow; Branch, Iterate and Jump. The most powerful of the three is Jump. In the early days of programming we used a dark art known as goto. Programming is more civilized now, instead we use the invocation of a function for that same behavior. 



## Branches: `if`, `elif`, `else`

```
import random
```

```
animal = random.choice(("Cat", "Dog", "Bug"))

if animal == "Cat":
    print("Mew")
elif animal == "Dog":
    print("Wuf")
else:
    print("Buz")
```

## Iteration: Loop Counter

```
counter = 0

while True:
    if counter >= 42:
        break
    else:
        counter += 1

print(counter)
```

## Range Based Iteration

```
for num in range(10):
    print(num * 10)    
```

## For Each Loop

```
some_list = [
    "Alpha",
    "Beta",
    "Gamma",
    "Delta",
    "Zeta",
]

for itm in some_list:
    print(itm)
```

## Enumeration

```
for idx, val in enumerate(some_list):
    print(idx, val)
```
